```python
from sklearn.datasets import load_iris
iris = load_iris()

X = iris.data
y = iris.target
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_new = pca.fit_transform(X)
```

```python
plt.title('IVE: {:2f}'.format(pca.explained_variance_ratio_.sum()))
plt.scatter(X_new[:,0], X_new[:,1], c=y, cmap=plt.cm.Set1,
            edgecolor='k')
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3).fit(X_new)
```

```python
plt.scatter(X_new[:, 0], X_new[:, 1], c=kmeans.labels_, cmap=plt.cm.Set1,
            edgecolor='k')
plt.show()
```
