```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
cols = ['country', 'description', 'designation', 'province', 'region_1', 'region_2', 'variety', 'winery']
X = df[cols].apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
X.head()
```

## Ajustando um modelo KNN

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
        X_counts, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier

mdl = RandomForestClassifier(n_estimators=10, n_jobs=-1)
mdl.fit(X_train,y_train)
y_pred = mdl.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

df_cm = pd.DataFrame(cm, index = y.unique(),
                  columns = y.unique())
plt.figure(figsize = (15,10))
sns.heatmap(df_cm.round(2), annot=True)
plt.show()
```
